```python
import pandas as pd
from src.preprocessing import preprocess
from src.train_models import train_logreg, train_rf, evaluate
from src.interpretation import logistic_coefficients, random_forest_importance

df = pd.read_csv("./data/german_credit_data_updated.csv")
df.head()
```

```python
df.shape
df.info()
df.describe(include="all")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Credit Risk", data=df)
plt.title("Target distribution (1 = good, 2 = bad)")
plt.show()
```

```python
X_train, X_test, y_train, y_test, preprocessor = preprocess(df)
```

```python
logreg = train_logreg(preprocessor)
logreg.fit(X_train, y_train)
evaluate(logreg, X_test, y_test, "Logistic Regression")
```

```python
rf = train_rf(preprocessor)
rf.fit(X_train, y_train)
evaluate(rf, X_test, y_test, "Random Forest")
```
